Total goal

```python
total_goal=df["Goals"].sum()
total_goal
```

Total Peanlty Goal

```python
total_penalty_goal=df["Penalty_Goals"].sum()
total_penalty_goal
```

```python
Total penalty attempted
```

```python
total_penalty_attempted=df["Penalty_Attempted"].sum()
total_penalty_attempted
```

```python
Conclusion:Out of 125 penalty attempts player where able to convert 102 into the goals
```

```python
Pie chart for penalty scored vs Penalty missed
```

```python
penalty_not_scored=total_penalty_attempted-total_penalty_goal
data=[total_penalty_goal,penalty_not_scored]
labels=["penalty scored","penalty missed"]
plt.pie(data,labels=labels,autopct='%1.2f%%')
plt.show()
```

Players based on age

```python
lessthan20=df[df["Age"]<20].count()
between20and25=df[df["Age"]>=20].count() & df[df["Age"]<25].count()
between25and30=df[df["Age"]>=25].count() & df[df["Age"]<30].count()
greaterthan30=df[df["Age"]>=30].count()
```

```python
data=[lessthan20["Age"],between20and25["Age"],between25and30["Age"],greaterthan30["Age"]]
label=["Age<20", "Age>20 and Age<=25", "Age>25 and Age<=30", "Age>30"]
plt.pie(data,labels=label,autopct='%.2f%%')
plt.title("player based on age")
plt.show()
```

```python
under20=df[df["Age"]<20]
under20["Club"].value_counts().plot(kind="bar")
plt.ylabel("Number of player")
```
